```python
from resume_ingest_docling import ingest_resume
resume_data = ingest_resume("nate_resume.docx")
```

```python
parsed_resume = resume_data
```

```python
import json
print(json.dumps(parsed_resume, indent=2))
```

```python
print("=== EDUCATION ===")
for edu in parsed_resume["education"]:
    print(edu)

print("\n=== PROJECTS ===")
for proj in parsed_resume["projects"]:
    print(proj["name"])
    for b in proj["bullets"]:
        print("  •", b)

print("\n=== EXPERIENCE ===")
for role in parsed_resume["experience"]:
    print(role["role_header"])
    for b in role["bullets"]:
        print("  •", b)

print("\n=== SKILLS ===")
for category, skills in parsed_resume["skills"].items():
    print(category, ":", skills)
```

```python
assert isinstance(parsed_resume["education"], list)
assert isinstance(parsed_resume["projects"], list)
assert isinstance(parsed_resume["experience"], list)
assert isinstance(parsed_resume["skills"], dict)

print("Schema validation passed.")
```

```python
import os
import getpass
from openai import OpenAI
from ai_resume_parser import AIResumeParser
from resume_ingest_docling import ingest_resume 

# 1. Initialize OpenAI Client securely
# Checks for environment variable first, otherwise prompts you
api_key = os.environ.get("OPENAI_API_KEY")
if not api_key:
    api_key = getpass.getpass("Enter OpenAI API Key: ")

client = OpenAI(api_key=api_key)

# 2. Ingest your resume
# Note: raw_data is the DICTIONARY returned, not the path strings
raw_data = ingest_resume("nate_resume.docx") 

# 3. Initialize Parser
parser = AIResumeParser(raw_data, llm_client=client, model_name="gpt-4o-mini")

# 4. Parse
result = parser.parse()
print(result["skills"]) 
```
